# Numerical instability example
Two nearly identical features ‚Üí linear dependence

```python
# ============================================================
# 1. LOAD DATA
# ============================================================

# Two features, x1 and x2, per sample y
x1 = np.array([1, 2, 3, 4, 5], dtype=float)
x2 = x1 + 1e-10  # almost identical

y = np.array([2, 4, 6, 8, 10], dtype=float)
```

```python
# ------------------------------------------------------------
# Pearson's Linear Correlation Coefficient
# ------------------------------------------------------------
# Compute Pearson correlation coefficient between x and y
# Using numpy built-in function
r = np.corrcoef(x1, y)[0, 1]

print("\nPearson correlation coefficient:")
print("r =", r)

# r ‚àà [-1, 1]:
#   r ‚âà 1   ‚Üí strong positive linear relationship
#   r ‚âà -1  ‚Üí strong negative linear relationship
#   r ‚âà 0   ‚Üí no linear correlation
```

```python
# ------------------------------------------------------------
# Matrix properties
# ------------------------------------------------------------
Phi = np.column_stack((x1, x2))
A = Phi.T @ Phi

print("\n--- Instability Example ---")
print("Œ¶^T Œ¶:")
print(A)

det_A = np.linalg.det(A)
print("Determinant:", det_A)

# Interpretation:
# det ‚âÝ 0  ‚Üí invertible
# det ‚âà 0  ‚Üí numerically unstable
# det = 0  ‚Üí singular (no solution via inverse)
```

```python
# ============================================================
# 2. NORMAL EQUATIONS (UNSTABLE)
# ============================================================
print("\n--- Normal Equations ---")

try:
    W_normal = np.linalg.inv(A) @ Phi.T @ y
    print("Weights (normal):", W_normal)
except np.linalg.LinAlgError as e:
    print("Inversion failed:", e)
    W_normal = None
```

```python
# ============================================================
# 3. SVD-BASED LEAST SQUARES (STABLE)
# ============================================================
print("\n--- Least Squares (SVD) ---")

W_lstsq, residuals, rank, s = '''to do'''

print("Weights (lstsq):", W_lstsq)
print("Rank:", rank)
print("Singular values:", s)
```

```python
# ============================================================
# 4. COMPARISON
# ============================================================
y_pred_lstsq = Phi @ W_lstsq

if W_normal is not None:
    y_pred_normal = Phi @ W_normal

plt.figure()
plt.scatter(x1, y, label="Data")

if W_normal is not None:
    plt.plot(x1, y_pred_normal, label="Normal equations (Fisher)")

plt.plot(x1, y_pred_lstsq, label="lstsq (SVD)")

plt.title("Numerical Stability Comparison")
plt.xlabel("x1 ‚âà x2")
plt.ylabel("y")
plt.legend()
plt.show()
```

KEY TAKEAWAYS:

1. Regression = optimization problem:
   We minimize sum of squared errors.

2. Normal equations:
   W = (Œ¶^T Œ¶)^(-1) Œ¶^T Y
   ‚Üí exact mathematical solution

3. Problem:
   If Œ¶^T Œ¶ is nearly singular:
   - determinant ‚âà 0
   - inversion becomes unstable

4. Why instability happens:
   - features are linearly dependent
   - small numerical errors get amplified

5. Better approach:
   - np.linalg.lstsq() (uses SVD)
   - avoids explicit matrix inversion
   - works even for rank-deficient systems

6. Practical rule:
   NEVER rely on explicit inverse in real systems
   ‚Üí use lstsq() or pinv()
